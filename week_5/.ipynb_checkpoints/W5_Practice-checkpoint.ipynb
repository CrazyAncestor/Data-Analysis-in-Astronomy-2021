```python
import numpy as np
import astropy.io.fits as pf
data = pf.open('new.fits')
```

```python
catalog = data[1].data

EBV = catalog['SFD']
HI = catalog['HI']/1e21
conversion_factor = 2*1e20/1e21
H2= catalog['CO10']*conversion_factor
```

```python
import matplotlib.pyplot as plt
```

```python

plt.plot(EBV,HI,label='Monte Carlo Method')
plt.plot(EBV,H2,label='Theoretical Expectation')
plt.xlabel('N')
plt.ylabel('error of mean')

plt.legend(loc='best')
plt.show()
```
